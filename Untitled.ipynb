#### Data cleaning <a class="anchor" id="section1_3"></a>
As data were analyzed it can be concluded there is no any outlier, but there are 2 missing values. So the unique action to do in the data cleaning is to drop the rows that contain the missing values. In adition, it would be necessary to create a new column called `income`, which is variable so important to analize.

```python
udemy_data = udemy_data.dropna()
```

```python
udemy_data["Income"] = udemy_data["price"]*udemy_data["num_subscribers"]
```

```python
udemy_data["creation_year"] = udemy_data["published_timestamp"].apply(lambda x: x.split("-")[0])
```

```python
udemy_data = udemy_data.reindex(columns=['course_id', 'course_title', 'url', 'price', 'num_subscribers', "Income",
       'num_reviews', 'num_lectures', 'level', 'Rating', 'content_duration',
       'published_timestamp', "creation_year", 'subject'])
#Convert creation_year to int and course_id to float
udemy_data.astype({"creation_year":"int", "course_id":"str"}).dtypes
```

### Exploratory Data Analysis <a class="anchor" id="EDA"></a>

After some data cleaning and tidying, the DataFrame is ready for EDA. The target is to answer the following questions:
* What is the relationship between rating and other variables such as content duration, number of lectures, etc?
* What is the relationship between published time and number of subscribers, and content duration and number of subscribers?
* What is the distributions of courses per level?
* How do the most profitable courses differ from the rest?

Along the way, this question may be refined and more questions may pop up.

#### Big picture <a class="anchor" id="section2_1"></a>

In order to observe the dataset as a whole, `DataFrame.hist()` is used.. It gives a full view of all numerical variables in the distribution. Most of the variables are skewed right and have similar distributions. Only `Rating` is skewed left.

Next, correlations between all numerical variables are viewed using a heat matrix. The heat matrix shows that `Rating`, one of the variables of most interest doesn't have any relationship with other variables. As we excepted, the most notorious relationships are between `Income` and `num_subscribers`; `num_lectures` and `num_reviews`.

```python
fig = plt.figure(figsize = (10,20))
ax = fig.gca()
udemy_data.hist(ax = ax);
```

```python
udemy_data.dtypes
```

```python
udemy_data.corr()
```

```python
plt.figure(figsize=(10,10))
# heat matrix that shows correlation across all numerical variables
sns.heatmap(data=udemy_data.corr(),annot=False)
plt.tight_layout()
plt.show()
```

#### Courses distribution per levels <a class="anchor" id="section2_2"></a>

In this section, the distribution of courses per level will be analyzed by creating bar and pie charts. These charts can be seen in the next lines.

```python
udemy_data.groupby("level").course_id.count().plot(kind="pie", autopct="%.1f%%")
plt.axis('Equal')
plt.title("Distribution of courses per level")
plt.tight_layout()
```

```python
sns.countplot(udemy_data.level, order=udemy_data.level.value_counts().index)
plt.xticks(rotation=40)
plt.title("Distribution of courses per level")
plt.show()
```

It can been seen that most of the courses are for all levels.

#### Income distribution <a class="anchor" id="section2_3"></a>
The objective of this section is to analyze how is of the distribution of the incomes get by the courses. For this, it will be necessary to create an histogram of the variable called `Income`.


```python
plt.hist(udemy_data.Income, bins=1000, range=(0,2000000))
plt.title("Income distribution")
plt.xlabel("Income (million dollars)")
plt.ylabel("Count")
plt.ylim([0,25])
plt.show()
```

From the histogram get, it can be seen that most of the courses have incomes between 10,000 y 100,000 dollars.

Altough, in the `Big picture` section wasn't found any interesting relationship between variables, in the next 2 sections it will try to see if there is any relationship between variables for some specific group of courses.

#### 20 hightest rating courses <a class="anchor" id="section2_4"></a>
In this section will be analyzed the relationship between all the variables for the 20 courses with the highest rating. To do this, first, the data will be sorted by the `Rating` column and then, stored in a variable called `top_20_rating`, and finally, it will be used the function `corr`, to see any dependency between variables.

```python
top_20_rating = udemy_data.sort_values(by=["Rating"], ascending=False).iloc[:20]
```

```python
top_20_rating.corr()
```

According to the results, there is no any relationship between `Rating` and other variables for the 20 courses with the highest rating.

#### 100 highest income courses <a class="anchor" id="section2_5"></a>
In this section will be analyzed the relationship between all the variables for the 100 courses with the highest income. To do this, first, the data will be sorted by the `Income` column and then, stored in a variable called `top_100_income`, and finally, correlations between all numerical variables are viewed using a heat matrix. In adition, it will know the courses distribution per level. 

```python
top_100_income = udemy_data.sort_values(by=["Income"], ascending=False).iloc[:100]
top_100_income.head(10)
```

```python
plt.figure(figsize=(10,10))
# heat matrix that shows correlation across all numerical variables
sns.heatmap(data=top_100_income.corr(),annot=False)
plt.tight_layout()
plt.show()
```

```python
top_100_income.corr()
```

```python
top_100_income.groupby("level").course_id.count().plot(kind="pie", autopct="%.1f%%")
plt.axis('Equal')
plt.title("Distribution of courses per level")
plt.tight_layout()
```

The heat matrix shows that `Income`,  doesn't have any relationship with other variables. Also, it was found that 50% of the courses with the highest income, are created for all levels, followed by the beginner level, with 35%.

#### Most popular courses <a class="anchor" id="section2_6"></a>
In this section will be analyzed the relationship between all the variables for the 20 courses with the highest number of subscribers. To do this, first, the data will be sorted by the `num_subscribers` column and then, stored in a variable called `most_polulars`, and then, it will be created the heat map, to see any dependency between variables. In adition, it will know the courses distribution per level.

```python
most_polulars = udemy_data.sort_values(by=["num_subscribers"], ascending=False).iloc[:20]
most_polulars
```

```python
plt.figure(figsize=(10,10))
# heat matrix that shows correlation across all numerical variables
sns.heatmap(data=most_polulars.corr(),annot=False)
plt.tight_layout()
plt.show()
```

```python
plt.hist(most_polulars.price, bins=100, range=(0,200))
plt.show()
```

```python
most_polulars.price.value_counts(normalize=True)
```

```python
most_polulars.groupby("level").course_id.count().plot(kind="pie", autopct="%.1f%%")
plt.axis('Equal')
plt.title("Distribution of courses per level")
plt.tight_layout()
```

The heat matrix shows that `num_subscribers`, doesn't have any relationship with other variables. Also, it was found that 50% of the courses with the highest income, are created for all levels, followed by the beginner level, with 35%.
